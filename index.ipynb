Base de datos: **world-happiness-report-2023.csv**

```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("happiness.csv")
```

Base de datos "cruda": ('head')

```python
df.head()
```

```python
columnas = [
    'Country name',
    'year',
    'Life Ladder',
    'Log GDP per capita',
    'Freedom to make life choices',
]

happy = df[columnas].copy()

happy = happy[happy['year'] >= 2022]
```

Nueva variable: Nivel de libertad para tomar decisiones

Creada a partir de la variable 'Freedom to make life choices'

```python
## FUNCIÓN -> Variable nueva

def clasificar_libertad(valor):
    if valor < 0.4:
        return "Baja libertad"
    elif valor < 0.7:
        return "Libertad media"
    else:
        return "Alta libertad"

happy['Freedom level'] = happy['Freedom to make life choices'].apply(
    clasificar_libertad
)
```

Base de datos modificada:

```python
happy.head()
```

```python
## Definir el tema

def para_graficos():
    return {
        "config": {
            "background": "#ffffff",
            "axis": {
                "labelFont": "Helvetica",
                "titleFont": "Helvetica",
                "labelFontSize": 11,
                "titleFontSize": 12
            },
            "legend": {
                "labelFont": "Helvetica",
                "titleFont": "Helvetica"
            },
            "title": {
                "font": "Helvetica",
                "fontSize": 14,
                "anchor": "start"
            }
        }
    }

alt.themes.register("para_graficos", para_graficos)
alt.themes.enable("para_graficos")
```

### **GRÁFICO 1: Relación entre el _Nivel de libertad para tomar decisiones_ y el _Life Ladder_**
Año 2022

```python
## GRÁFICO 1: Freedom level vs Life Ladder

grafico_1 = alt.Chart(happy).mark_point(
    opacity=0.5
).encode(
    x=alt.X(
        'Freedom level:N',
        title='Nivel de libertad para tomar decisiones'
    ),
    y=alt.Y(
        'Life Ladder:Q',
        title='Life Ladder'
    ),
    color='Freedom level:N',
    tooltip=['Country name', 'Life Ladder']
).properties(
    width=200,
    height=400,
    title={
        "text": "Felicidad según nivel de libertad percibida",
        "subtitle": [
            "Fuente: World Happiness Report"
        ]
    }
)

grafico_1
```

### **GRÁFICO 2: Relación entre el _PBI per cápita_ y la _Libertad para tomar decisiones_**
Año 2022

```python
## GRÁFICO 2: Log GDP per capita vs Freedom to make life choices

grafico_2 = alt.Chart(happy).mark_point(
    opacity=0.6
).encode(
    x=alt.X(
        'Freedom to make life choices:Q',
        title='Libertad para tomar decisiones'
    ),
    y=alt.Y(
        'Log GDP per capita:Q',
        title='PBI per cápita'
    ),
    color='Freedom level:N',
    tooltip=[
        'Country name',
        'Freedom to make life choices',
        'Log GDP per capita'
    ]
).properties(
    width=300,
    height=300,
    title={
        "text": "PBI per cápita y libertad para tomar decisiones",
        "subtitle": [
            "Fuente: World Happiness Report"
        ]
    }
)

grafico_2
```
